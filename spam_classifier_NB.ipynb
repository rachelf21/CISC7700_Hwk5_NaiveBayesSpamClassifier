```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from mpl_toolkits.axes_grid1 import AxesGrid
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

dataOriginal = pd.read_csv("spambase_data.csv")
dataset = np.array(dataOriginal)
features = dataset[:,0:48]
labels = dataset[:,-1]
total_observations = len(dataset)
x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=.3, random_state=15)
```

```python
# %% Multinomial Naive Bayes Model

model_MultiNB = MultinomialNB()
model_MultiNB.fit(x_train, y_train)
y_pred = model_MultiNB.predict(x_test)
print("\n", model_MultiNB)
print("Accuracy Score: " , accuracy_score(y_test, y_pred))

conf_matrix = confusion_matrix(y_test,y_pred)
title = "\nConfusion Matrix for Multinomial Naive Bayes Classifier\n"
print(title, conf_matrix, "\n")

fig, ax = plt.subplots(figsize=(5,5))
sns.heatmap(conf_matrix, annot=True, annot_kws={"size": 24}, fmt='d', cmap="Blues", square=True, cbar=False, xticklabels =["Spam","Not Spam"],yticklabels=["Spam","Not Spam"])
ax.tick_params(labelsize=12)
ax.set_title("Confusion Matrix for \nMultinomial Naive Bayes Classifier", fontsize=14, weight='bold')
ax.set_xlabel("Predicted", fontsize=14, labelpad=20, weight='bold')
ax.set_ylabel("Actual", fontsize=14, labelpad=20, weight='bold')

conf_matrix_norm = conf_matrix/(total_observations*.3)
title = "Normalized Confusion Matrix for Multinomial Naive Bayes Classifier\n"
print(title, conf_matrix_norm)

fig, ax = plt.subplots(figsize=(5,5))
sns.heatmap(conf_matrix_norm, annot=True, annot_kws={"size": 24}, fmt='.0%', cmap="Blues", square=True, cbar=False, xticklabels =["Spam","Not Spam"],yticklabels=["Spam","Not Spam"])
ax.tick_params(labelsize=12)
ax.set_title("Normalized Confusion Matrix for \nMultinomial Naive Bayes Classifier", fontsize=14, weight='bold')
ax.set_xlabel("Predicted", fontsize=14, labelpad=20, weight='bold')
ax.set_ylabel("Actual", fontsize=14, labelpad=20, weight='bold')

plt.show()
accuracy = np.round(accuracy_score(y_test, y_pred),3)
print("Accuracy Score for Multinomial Naive Bayes Classifier (Test size=30%): {0:.1%}".format(accuracy))
```
